```python
# Generate 10000 random numbers with normal distribution

a = np.arange(10000)
l = []
for i in a:
    l.append(np.random.randn())

# calculate and print mean and std. dev.
std = np.std(l)
mean = np.mean(l)
print(f'Mean = {mean}')
print(f'Standard Dev. = {std}')

# plt a histogram with 30 bins
plt.hist(l, bins=30);

```
